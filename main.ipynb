```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime = pd.read_csv("crime_dataset_india.csv")
```

```python
crime.head()
```

```python
crime_city = crime.groupby('City')['Report Number'].count().reset_index().sort_values(by = 'Report Number', ascending = False)
crime_city
```

```python
crime['Year'] = pd.to_datetime(crime['Date of Occurrence'], errors = 'coerce').dt.year
```

```python
most_crime_year = crime.groupby('Year')['Report Number'].count().reset_index().sort_values(by = 'Report Number', ascending = False)
most_crime_year
```

```python
delhi_crime = crime[crime['City'] == 'Delhi']
delhi_crime_rate = delhi_crime.groupby('Year')['Report Number'].count().reset_index()
delhi_crime_rate
```

```python
crime_type = crime.groupby('Crime Description')['Report Number'].count().reset_index().sort_values(by = 'Report Number', ascending=False)
crime_type
```

```python
victim_gender = crime.groupby('Victim Gender')['Report Number'].count().reset_index().sort_values(by = 'Report Number', ascending=False)
victim_gender
```

```python
crime['Month'] = pd.to_datetime(crime['Date of Occurrence']).dt.month
```

```python
lt_2024 = crime[crime['Year'] < 2024]
crime_month = lt_2024.groupby('Month')['Report Number'].count().reset_index()
crime_month
```

```python
avg_age_of_victims = np.mean(crime['Victim Age'], axis = 0)
print(f"Average age of victims (2020–2024): {avg_age_of_victims:.2f} years")
```

```python
female_victim = crime[crime['Victim Gender'] == 'F']
female_avg_age = np.mean(female_victim['Victim Age'])
print(f"Average age of female victims (2020–2024): {female_avg_age:.2f} years")
```

```python
male_victim = crime[crime['Victim Gender'] == 'M']
male_avg_age = np.mean(male_victim['Victim Age'])
print(f"Average age of female victims (2020–2024): {male_avg_age:.2f} years")
```
